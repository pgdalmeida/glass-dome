# QUESTION: Is each activity (act_id) unique to each MP (mp_id/mp_name)? 
No, but almost. For roughly half of the MPs, the act_id's are unique, and for the remainder there's a maximum of 8 duplicates for MP PEDRO DO CARMO, out of 314 activities.

```python
act_counts_df = mp_act_df.groupby('mp_name')['act_id'].agg(['count', 'nunique'])
act_counts_df['diff'] = act_counts_df['count'] - act_counts_df['nunique']
act_counts_df.query('diff > 0').sort_values('diff')
```

```python
file_path = '/home/pedro/Projects/glass-dome/csv_data/atividade_dos_deputados/xv_legislatura/atividadedeputadoxv.csv'
mp_act_df = pd.read_csv(file_path, sep='|')
mp_act_merged_df = mp_act_df.merge(mp_act_df, on=['act_type', 'act_subtype', 'act_id'], suffixes=('', '_coauthors') )
n_authors = mp_act_merged_df.groupby(['mp_name', 'act_id'], as_index=False)['mp_name_coauthors'].agg('nunique')
df = mp_act_df.merge(n_authors, on=['mp_name', 'act_id'])
df['act_weight'] = 1/df['mp_name_coauthors']
df = df[['mp_id', 'mp_name', 'mp_party', 'act_id', 'act_type', 'act_subtype', 'act_weight']].drop_duplicates()
df.groupby(['mp_party', 'mp_name'])['act_weight'].sum().sort_values()
```
