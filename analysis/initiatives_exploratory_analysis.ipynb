```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = '/home/pedro/Projects/glass-dome/csv_data/iniciativas/iniciativasxv.csv'
ini_df = pd.read_csv(file_path, sep='|')
```

#### Histogram the number of authors per initiative

Histograming the number of mps envolved in each initiative, as to understand how independent is the work of each mp when compared to the rest of the party. If all the mps sign every initiative of the party together every time, it becomes impossible to distinguish each mp's contribution to the parliamentary activity. 

```python
hist_df = ini_df.groupby(['party', 'number_authors'])['id'].count().reset_index(name='count')
n_bins = 100
fig, axs = plt.subplots(nrows=4, ncols=2, sharey=True)
fig.tight_layout(h_pad=2)

party_list = {
    'BE':   (3,0), 
    'PCP':  (2,0), 
    'PAN':  (1,0), 
    'L':    (0,0), 
    'PS':   (0,1), 
    'PSD':  (1,1),
    'IL':   (2,1),
    'CH':   (3,1) 
}
for p, (x,y) in party_list.items():
    axs[x, y].set_title(p)
    axs[x, y].hist(ini_df.query(f'party == "{p}"')['number_authors'])



```
